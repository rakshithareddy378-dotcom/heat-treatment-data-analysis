# Effect of Heat Treatment Temperature ,Cooling Methods and Impact energy on Hardness


## Objective
The objective of this project is to analyze the effect of heat treatment temperature,cooling method and Impact energy using Python-based data analysis and visualization techniques.


## Dataset Description
The dataset used in this project represents heat treatment process parameters and their corresponding mechanical properties.  
It includes the following variables:

- Heat treatment temperature (Â°C)
- Cooling method (Air, Oil, Water)
- Hardness (HRC)
- Impact energy (J)

This dataset is used to study the relationship between heat treatment parameters and mechanical properties.


## Data Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\Rakshitha Reddy\\Downloads\\heat_treatment_data.csv")
cooling_hardness = df.groupby("Cooling_Method")["Hardness_HRC"].mean()
cooling_hardness

cooling_hardness.plot(kind="bar")
plt.title("Average Hardness for Different Cooling Methods")
plt.xlabel("Cooling Method")
plt.ylabel("Hardness (HRC)")
plt.show()



```

**Observation:**  
The bar chart shows that water quenching results in the highest average hardness, followed by oil and air cooling. This is due to the higher cooling rate of water, which promotes martensite formation.



```python
# Create temperature ranges
df["Temp_Range"] = pd.cut( df["Heat_Temp"], bins=[750, 800, 850, 900, 950],labels=["750â800", "800â850", "850â900", "900â950"])

# Calculate average hardness for each temperature range
temp_hardness = df.groupby("Temp_Range", observed=True)["Hardness_HRC"].mean()
temp_hardness

plt.figure()
plt.plot(temp_hardness.index, temp_hardness.values, marker='o')
plt.title("Effect of Heat Treatment Temperature on Hardness")
plt.xlabel("Heat Treatment Temperature Range (Â°C)")
plt.ylabel("Average Hardness (HRC)")
plt.show()
```

### Observation
From the graph, it is observed that heat treatment temperature has a significant effect on hardness. Moderate temperature ranges result in higher hardness, while very high temperatures do not further improve hardness and may slightly reduce it.


```python
plt.figure()
plt.scatter(df["Hardness_HRC"], df["Impact_Energy_J"])
plt.title("Hardness vs Impact Energy")
plt.xlabel("Hardness (HRC)")
plt.ylabel("Impact Energy (J)")
plt.show()
```

```python
Observation:
The graph shows that an increase in hardness leads to a decrease in impact energy, indicating a trade-off between hardness and toughness.
```

```python
corr = df[["Heat_Temp", "Hardness_HRC", "Impact_Energy_J"]].corr()
corr
```

##Conclusion
- Heat treatment temperature has a significant influence on hardness, with moderate temperature ranges producing higher hardness values.
- Cooling method affects hardness due to differences in cooling rate, with faster cooling generally resulting in higher hardness.
- An inverse relationship is observed between hardness and impact energy, indicating a clear strengthâtoughness trade-off.
